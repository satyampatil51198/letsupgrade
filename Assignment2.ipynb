# Report card

```python
percentage = int(input("Enter your percentages"))

if(percentage <= 50):
    print("Grade D")
elif((percentage > 50)and(percentage <= 65)):
    print("Grade C")
elif((percentage > 65)and(percentage <= 80)):
    print("Greade B")
elif((percentage > 80)and(percentage <= 100)):
    print("Grade A")
```

# prime number 1 to 1000

```python
for n in range(1,1001):
    count = 0
    for i in range(2,n):
        if(n % i == 0):
            count=count+1
            break
    if(count == 0):
            print(n)
```
